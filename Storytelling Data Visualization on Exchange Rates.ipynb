# Storytelling Data Visualization on Euro-USD Exchange Rates

## Introduction

Our last project offered a chance to display our newfound skills in exploratory data visualization. For this project, we are focusing on explanatory data visualization — creating graphs for others in order to inform, make a point, or tell a story.

The story we'll tell today is of the Euro exchange rates against the US dollar during the 2007-08 financial crisis. The euro, symbolized with €, is the official currency in most of the countries of the European Union. For instance, if the exchange rate of the euro to the US dollar is 1.5, we get 1.5 US dollars if we pay 1.0 euro. (One euro has more value than one dollar at this exchange rate.)

Daria Chemkaeva put together the dataset we'll be using and made it available on [Kaggle](https://www.kaggle.com/datasets/lsind18/euro-exchange-daily-rates-19992020). This is a much larger dataset and contains daily Euro exchange rate data from 1999 to the present day. The data source is the European Central Bank. Note that the dataset gets regular updates every few months; we'll be working with a version from January 2021.

## Getting Started

```python
import pandas as pd
exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
exchange_rates.head()
```

```python
exchange_rates.tail()
```

```python
exchange_rates.info()
```

Our dataset has 5699 rows and 41 columns; the first column includes a date and the other 40 are for various world currencies. Each row represents a day.

Several columns have null values. Some are for an obvious reason — the currency the column represents became obsolete; one example is the Greek drachma which experienced this fate when Greece joined the Eurozone in 2001, two years after the beginning of data collection.

In a dataset full of exchange rates, one would expect the values to be float types, but there are only three such columns in ours. The rest are reported as strings, but pandas will also list an entire column as `object` type if it has null values, for example. The first column is also full of strings when we would ideally want datetime objects. In any case, we'll have some converting to do.

## Data Cleaning

Before we start creating data visualizations, we'll need to do a bit of data cleaning. Our focus for this project will be on the exchange rate between the euro and the US dollar. So, we will do the following:

- Rename the `[US dollar ]` and `Period\Unit:` columns to something easier to type — `US_dollar` and `Time`.
- Change the `Time` column to a `datetime` data type.
- Sort the values by `Time` in ascending order.
- Reset the index (and drop the initial index).

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
```

Now we're going to isolate the `Time` and `US dollar` columns and assign them to their own DataFrame. If we want to explore a different currency afterwards, the code in this section will come in handy.

```python
euro_to_dollar = pd.DataFrame(exchange_rates[['Time', 'US_dollar']]).copy()
euro_to_dollar['US_dollar'].value_counts()
```

There are 62 rows in the `US_dollar` column with a `-` value. This is not useful for visualization, so we will drop these rows. We will also convert the rest of the values to the `float` type.

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
```

## Rolling Mean

Now that we're finished cleaning the data, we'll generate a line plot to visualize the evolution of the euro-dollar exchange rate.

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
plt.show()
```

If we look at the line's shape, we see many small wiggles rather than a smooth line. The wiggles, however have meaning — they are the visual representation of the daily variation in the exchange rate. The rate goes up and down from day to day. We will only see clearer upward or downward trends in the longer run (months or years).

Depending on our goals, we may not want to show that daily variation on our graph. If we want to hide it and show only the long-term trends, we can use the **rolling mean**. We can calculate the mean for each day using the `pandas.Series.rolling().mean()` method. One of the parameters this method takes is the **rolling window**, which is the number of days before the current day used in the calculation. If the rolling window is three, then we calculate the mean between the value of the current day and the values of the previous two days.

Let's try this out by calculating the rolling means for the `US_dollar` column using a window of 30 days.

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar[28:]
```

Looks good. We should expect the first 29 rows to have a `NaN` value.

## Sketching a Graph

Now that we know what story we want to tell, we're going to think more about how our graph is going to look. As a general practice, it's a good idea to sketch the graph on pencil (or pen) and paper, or even a graphics editor, before coding it.

For ours, it will take the form of a line plot between the beginning of 2006 through the end of 2009. The years 2007 and 2008 will be bolded and differentiated by color. We will most likely design the graph in the FiveThirtyEight style.

## Coding the Graph

Below we will begin coding our graph. After we display it, we will explain our data presentation choices as they reflect the design principles we've learned about.

```python
# Isolate the relevant data first

financial_crisis = euro_to_dollar.copy(
                   )[(euro_to_dollar['Time'].dt.year >= 2006
                   ) & (euro_to_dollar['Time'].dt.year <= 2009)]
financial_crisis_main = euro_to_dollar.copy(
                   )[(euro_to_dollar['Time'].dt.year >= 2007
                   ) & (euro_to_dollar['Time'].dt.year <= 2008)]
```

```python
# Add the FiveThirtyEight style
import matplotlib.style as style
style.use('fivethirtyeight')

# Add the basic plot
fig, ax = plt.subplots(figsize=(8,5))
ax.plot(financial_crisis['Time'], financial_crisis['rolling_mean'],
        linewidth=1.5, color='#A6D785')

# Add the financial crisis plot
ax.plot(financial_crisis_main['Time'], financial_crisis_main['rolling_mean'],
        linewidth=2.5, color='#e23d28')

# Remove default tick labels
ax.set_xticklabels([])
ax.set_yticklabels([])

# Rewrite and position labels
# Positioning takes so much trial and error!
# Using `ax.get_xticks()` and `ax.get_yticks()` is helpful here
starting_x = 13105.0
for year in [2006, 2007, 2008, 2009, 2010]:
    ax.text(starting_x, 1.14, year, alpha=0.5, fontsize=11)
    starting_x += 365
    
starting_y = 1.195
for rate in [1.2, 1.3, 1.4, 1.5]:
    ax.text(13000.0, starting_y, rate, alpha=0.5, fontsize=11)
    starting_y += 0.1
    
#Highlight the peak of the crisis
ax.axvspan(xmin=13940.0, xmax=14140.0, alpha=0.3, color='grey')
    
# Add title and subtitle
ax.text(13070.0, 1.65, 'Euro-USD rate peaked at 1.59 during 2007-08 financial crisis', weight='bold')
ax.text(13070.0, 1.63, 'Euro-USD exchange rates from 2006-2010', size=12)

# Add signature
ax.text(13000, 1.08, ' '*80 + 'Source: European Central Bank', size=12,
        color = '#4d4d4d', weight='bold')

# Lighten grid
ax.grid(alpha=0.5)

plt.show()
```

There are several design principles guiding this graph's aesthetic. Here we:

- Maximized the data-ink ratio by:
    - Removing all tick lines
    - Labeling every other tick
    - Including key context and takeaway in the title and subtitle
    - Lightening the background grid
- Differentiated the crisis years with the use of pre-attentive attributes:
    - Color (red)
    - Size/thickness
- Highlighted the peak of the crisis via the enclosure principle, also a pre-attentive attribute
- Separated the signature from the rest of the graph via enclosure
- Presented the data through a line graph, a widely recognized format, via the familiarity principle
